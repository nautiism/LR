```python
#data prep
var = ['mainroad','guestroom','basement','airconditioning','hotwaterheating','prefarea']

house[var] = house[var].apply(lambda x : x.map({'yes':1,'no':0}))
```

```python
house.head()
```

```python
make = pd.get_dummies(house['furnishingstatus'],drop_first=True)
house = pd.concat([house,make],axis=1)
house.drop(['furnishingstatus'],axis=1,inplace=True)
house.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
```

```python
num_variables = ['area','bedrooms','bathrooms','stories','parking','price']
house[num_variables] = scalar.fit_transform(house[num_variables])
```

```python
house.head()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(house.corr(),annot=True,cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(house.area,house.price)
plt.show()
```

```python
Y_train = house.pop('price')
X_train = house
```

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train['area'])
lr = sm.OLS(Y_train,X_train_lm).fit()
lr.params
```

```python
plt.scatter(X_train_lm.iloc[:,1],Y_train)
plt.plot(X_train_lm.iloc[:,1],0.121175*X_train_lm.iloc[:,1]+0.581968,'r')
plt.show()
```

```python
print(lr.summary())
```

```python
#2nd higherst correlation , bathroom
X_train_lm = sm.add_constant(X_train[['area','bathrooms']])
lr = sm.OLS(Y_train,X_train_lm).fit()
lr.params
```

```python
print(lr.summary())
```

```python
#all variables 
X_train_lm = sm.add_constant(X_train)
lr = sm.OLS(Y_train,X_train_lm).fit()
lr.params
```

```python
lr.summary()
```

```python
#Check VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
vif = pd.DataFrame()
vif['features'] = house.columns
vif['VIF'] = [ VIF(house.values,i) for i in range(house.shape[1]) ]
vif['VIF'] = round(vif['VIF'],2)
vif.sort_values(by="VIF",ascending=False)
```

```python
#throw away bedrooms
#all variables 
X_train = X_train.drop(['bedrooms'],axis=1)
X_train_lm = sm.add_constant(X_train)
lr = sm.OLS(Y_train,X_train_lm).fit()
lr.params
```

```python
lr.summary()
```

```python
y_hat = lr.predict(X_train_lm)
res = Y_train-y_hat
```

```python
res.head()
```

```python
#error distribution
fig = plt.figure()
sns.distplot(res,bins=15)
plt.show()
```
